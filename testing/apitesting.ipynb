```python
import requests

api_key = "d376tm1r01qtvbtkcj00d376tm1r01qtvbtkcj0g"
url = f"https://finnhub.io/api/v1/stock/symbol?exchange=US&token={api_key}"

response = requests.get(url)
if response.status_code == 200:
    company_list = response.json()
    
    # Extract just the company names into a list
    company_names = [company['description'] for company in company_list]
    
    print(f"Total companies found: {len(company_names)}")
    print("Sample companies:", company_names) 
else:
    print("Failed to fetch data, status code:", response.status_code)
```
